```python
#Import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

2.Load the dataset using pandas read_csv() function


```python
#Load cars.csv dataset
car=pd.read_csv('/content/gdrive/MyDrive/principle of datascience/cars.csv')
car
```

3.Display the head of the dataset using head() function

```python
#display the head of the dataset
car.head()
```

4.Display the bottom 5 row from the dataset using tail() function

```python
#Display the bottom 5 rows
car.tail(5)
```

5.Print summary statistical of data set using the describe() function

```python
#Analytical summary of dataset
car.describe(include='all')
```

6.plot a histogram for all the variables

```python
car.hist(figsize=(20,30))
```

7.Box plot to vizualize the relationship between model and size

```python
#Relationship boxdescion categorical nd continuous vertical
sns.boxplot(x="Size",y="CITY (kWh/100 km)",data=car)
```

8.Build a pairplot using subborn library


```python
sns.pairplot(car)
```

9.Drop irrelevent columns from the datset using drop() function

```python
#drop irrelevent column
car=car.drop(['YEAR','RATING'],axis=1)
car.head(5)
```

10.use rename() function to rename


```python
#Rename column name
car=car.rename(columns={"Model":"Vehicle_Model","Size":"Vehicle_size"})
car.head(5)
```

11.Print the total number of duplicate row

```python
#rows containing duplicate data
duplicate_rows_car=car[car.duplicated()]
print("Number of duplicate rows:",duplicate_rows_car.shape)
```

```python
#count the rows before removing the data
car.count()
```

12.Remove the duplicate rows using drop_duplicates () function

```python
#drop the duplicates
car.drop_duplicates()
car.head()
```

13.Drop the missing values from th dataset

```python
#find the null values
print(car.isnull().sum())
```

```python
#drop the missing value
car=car.dropna()
car.count()
```

14.Plot a histogram to find the number of cars per brand

```python
car.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number of cars by Make")
plt.ylabel('Nuber of cars')
plt.xlabel('Make')
```

15.Draw a correlation plot between the variables


```python
#find the relation between the value
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
